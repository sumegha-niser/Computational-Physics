```python
import numpy as np
from library import read_data

#reading the data
data = read_data("assign3fit.txt")

xlist = data[:, 0]
ylist = data[:, 1]
```

```python
import numpy as np
from library import polynomial_fit

coefficients = polynomial_fit(xlist, ylist, np.ones(len(xlist)), 3)
print("Coefficients are:", coefficients)
```

```python
def evaluate_polynomial(coefficients, x):
    return sum([coefficients[i] * x**i for i in range(len(coefficients))])

import matplotlib.pyplot as plt
import numpy as np

def plot_data_and_fit(x, y, coefficients):
    # Create a new figure with a specified size
    plt.figure(figsize=(13, 6))

    # Plot the given data points
    plt.scatter(x, y, color='black', label='Given Data points')

    # Generate a range of x-values for the fitted curve
    x_fit = np.linspace(min(x), max(x), 1000)

    # Evaluate the fitted polynomial at the x-values
    y_fit = evaluate_polynomial(coefficients, x_fit)

    # Plot the fitted curve
    plt.plot(x_fit, y_fit, color='blue', label='Fitted Curve')

    # Set the labels and title of the plot
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Fitting the data with polynomial of degree 3')

    # Add a grid and a legend
    plt.grid(True)
    plt.legend()

    # Display the plot
    plt.show()

# Call the function with your data and coefficients
plot_data_and_fit(xlist, ylist, coefficients)
```
